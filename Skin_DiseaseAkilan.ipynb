```python

def plot_prc():
    for i in range(1, 7):   
        precision, recall, thresholds = prc[i]

        plt.plot(recall,precision, label=f"Class {i} PR curve")

        # find threshold closest to zero
        close_zero = np.argmin(np.abs(thresholds))

        plt.plot(
            recall[close_zero],
            precision[close_zero],
            'o',
            markersize=8,
            label=f"Class {i} threshold≈0")

        print(f"Class {i}: threshold closest to zero = {thresholds[close_zero]}")

        plt.xlabel("Recall")
        plt.ylabel("Precision")
        plt.legend()
        plt.show()
```

```python
plt.figure(figsize=(8,6))

for i in roc_dict.keys():
    fpr, tpr, roc_auc = roc_dict[i]
    plt.plot(fpr, tpr, label=f"Class {i} (AUC={roc_auc:.2f})")

plt.plot([0,1], [0,1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (One-vs-Rest)")
plt.legend()
plt.show()
```
